### Define Performance Metric Function

This code block defines a function to evaluate the performance of a model using the $R^2$(coefficient of determination) metric. The steps are as follows:

1. **Imported Function:**
   - `r2_score` from `sklearn.metrics`: Calculates the $R^2$value, which measures the proportion of variance in the dependent variable that is predictable from the independent variables.

2. **Function Definition:**
   - `performance_metric(y_true, y_predict)`: Takes two arguments:
     - `y_true`: The actual target values.
     - `y_predict`: The predicted target values.
   - Returns the $R^2$score as `score`.

3. **Purpose:**
   - $R^2$ranges from 0 to 1, where higher values indicate better model performance.
   - This function will be used to evaluate how well the model's predictions match the actual data.

This step establishes a standardized way to assess model accuracy during the training and testing phases.

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
  score = r2_score(y_true, y_predict)
  return score
```

### Evaluate Model Performance Using $R^2$Metric

This code block evaluates the performance of a model by calculating its $R^2$(coefficient of determination) score. The following steps are performed:

1. **Input Data:**
   - `y_true`: The actual target values, provided as `[3, -0.5, 2, 7, 4.2]`.
   - `y_predict`: The predicted target values, provided as `[2.5, 0.0, 2.1, 7.8, 5.3]`.

2. **Function Call:**
   - The `performance_metric` function, defined earlier, is called with `y_true` and `y_predict` as inputs.
   - The function computes the $R^2$score using `sklearn.metrics.r2_score`.

3. **Output:**
   - The calculated $R^2$score is printed in a formatted message, showing the model's ability to explain the variance in the data.

This step demonstrates the model's performance by quantifying how well its predictions align with the actual values.

```python
# Calculate the performance of this model
score = performance_metric([3, -0.5, 2, 7, 4.2], [2.5, 0.0, 2.1, 7.8, 5.3])
print("Model has a coefficient of determination, R^2, of {:.3f}.".format(score))
```

### Define Model Fitting Function with Grid Search

This code block defines a function to fit a Decision Tree Regressor model to the dataset using grid search for hyperparameter tuning. Specifically, it tunes the `max_depth` parameter of the decision tree. Here's what the code does:

1. **Function Definition:**
   - `fit_model(X, y)`: This function accepts two parameters:
     - `X`: The input features.
     - `y`: The target variable.
   - The function performs grid search over the `max_depth` hyperparameter of a Decision Tree Regressor.

2. **Cross-Validation Setup:**
   - `ShuffleSplit(X.shape[0], test_size=0.20, random_state=0)`: Creates cross-validation sets by randomly splitting the data into training and testing subsets, with 20% allocated for testing.
   
3. **Grid Search:**
   - `DecisionTreeRegressor()`: Initializes the decision tree regressor.
   - `params`: Defines the hyperparameter grid for `max_depth` ranging from 1 to 10.
   - `make_scorer(performance_metric)`: Uses the previously defined performance metric to evaluate the model.
   - `GridSearchCV`: Performs grid search to find the best `max_depth` value using cross-validation.

4. **Output:**
   - Returns the best model after fitting the data, based on the highest $R^2$score.

This step is essential for tuning the model's complexity and selecting the optimal decision tree depth to improve performance.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a
        decision tree regressor trained on the input data [X, y]. """

    # Create cross-validation sets from the training data
    cv_sets = ShuffleSplit(X.shape[0], test_size = 0.20, random_state = 0)

    regressor = DecisionTreeRegressor()
    params = {'max_depth': range(1,11)}
    scoring_fnc = make_scorer(performance_metric)
    grid = GridSearchCV(regressor,params,scoring=scoring_fnc,cv=cv_sets)
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    return grid.best_estimator_
```

### Fit Model to Training Data and Identify Optimal `max_depth`

This code block fits the training data to the decision tree model using grid search and retrieves the optimal `max_depth` parameter. Here's what the code does:

1. **Fit the Model:**
   - `reg = fit_model(X_train, y_train)`:
     - The `fit_model` function, defined earlier, is called to fit the training data (`X_train`, `y_train`) to a Decision Tree Regressor using grid search.
     - The optimal model is returned and stored in the variable `reg`.

2. **Extract Optimal Hyperparameter:**
   - `reg.get_params()['max_depth']`:
     - Retrieves the optimal value of the `max_depth` hyperparameter found during the grid search.

3. **Output:**
   - Prints the optimal `max_depth` parameter for the trained model.

This step allows for tuning the model by identifying the best value for the `max_depth` parameter, ensuring the decision tree model is well-optimized for the dataset.

```python
# Fit the training data to the model using grid search
reg = fit_model(X_train, y_train)

# Produce the value for 'max_depth'
print("Parameter 'max_depth' is {} for the optimal model.".format(reg.get_params()['max_depth']))
```
