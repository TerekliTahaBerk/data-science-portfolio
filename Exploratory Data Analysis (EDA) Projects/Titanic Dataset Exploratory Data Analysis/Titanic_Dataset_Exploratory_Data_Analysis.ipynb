```python
# Improved Countplot for Pclass by Sex
plt.figure(figsize=(10, 6))

# Create the countplot with enhanced aesthetics
sns.catplot(
    x='Pclass',
    data=titanic_df,
    hue='Sex',
    kind='count',
    palette='Set2',  # Softer color palette for better visuals
    height=6,
    aspect=1.5
)

# Add title and axis labels
plt.title('Passenger Class Distribution by Sex', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Passenger Class (Pclass)', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Count of Passengers', fontsize=14, fontweight='bold', color='darkblue')

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for better presentation
plt.tight_layout()

# Show plot
plt.show()
```

Another thing we can do before we start infering something from the above factorplot, is to divide the passengers between male, female, and a child.

```python
#Function to detect if a person is a man, woman or child.
def man_wom_chi(passenger):
    age=passenger['Age']
    sex=passenger['Sex']

    return 'child' if age < 16 else sex

#Using Pandas' apply method to create a new column "Person"
titanic_df['Person'] = titanic_df.apply(man_wom_chi,axis=1)
```

## What deck were the passengers on and how does that relate to their class?


As we're doing open ended analysis, let's try and check out if people belonging to higher classes were assigned cabins on a higher deck (or level) of the ship.

```python
#Because the cabin data was missing in a lot of cases, we can just drop it for now for this section.
deck_df = titanic_df.dropna(axis=0)
```

```python
deck_df.head()
```

How can we determine which deck a passenger was assigned to?

The **Cabin** attribute contains this information. Typically, a passengerâ€™s cabin number combines their deck letter and room number. For example, a passenger on deck â€˜Câ€™ would have a cabin number in the format â€˜CXXX,â€™ where â€˜XXXâ€™ represents the room number.

To extract the deck information, we can simply create a Python method to retrieve the first character from the cabin data.

```python
#Grabbing the deck from the cabin numbers
def get_level(passenger):
    cabin = passenger['Cabin']
    return cabin[0]

#  get_level[deck_df.iloc[1]]
```

```python
deck_df['level']=deck_df.apply(get_level,axis=1)
```

```python
deck_df.head()
```

Everything seems to work, so now we can check out how many passengers belonged to different decks.



```python
# Improved Countplot for 'level' in deck_df
plt.figure(figsize=(10, 6))

# Create the countplot with enhanced aesthetics
sns.catplot(
    x='level',
    data=deck_df,
    kind='count',
    palette='coolwarm',  # Softer color palette for better visuals
    height=6,
    aspect=1.5
)

# Add title and axis labels
plt.title('Distribution of Levels', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Level', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='darkblue')

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for better presentation
plt.tight_layout()

# Show plot
plt.show()
```

And to address our original questionâ€¦

```python
# Improved Countplot for 'level' by 'Pclass' in deck_df
plt.figure(figsize=(12, 6))

# Create the countplot with enhanced aesthetics
sns.catplot(
    x='level',
    data=deck_df,
    hue='Pclass',
    kind='count',
    palette='coolwarm',  # Softer color palette for better visuals
    height=6,
    aspect=1.5
)

# Add title and axis labels
plt.title('Distribution of Levels by Pclass', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Level', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='darkblue')

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for better presentation
plt.tight_layout()

# Show plot
plt.show()
```

It appears that passengers from the upper classes were indeed assigned to the upper decks. Decks A, B, and C were exclusively allocated to first-class passengers, while third-class passengers only begin to appear from deck E onwards. However, itâ€™s important to note that cabin information is missing for more than three-quarters of the passengers, which limits our analysis.

## Who was with their family?


Letâ€™s revisit our data for a closer examination:

```python
titanic_df.head()
```

The **SibsSP** attribute indicates the number of siblings or spouses a passenger had aboard, while Parch represents the number of parents or children traveling with them.

Since weâ€™re mainly interested in whether a passenger had any family members onboard, we can simplify our analysis by creating a new column to represent this information.

```python
#Adding the number of family a passenger had onboard
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
titanic_df.tail()
```

Any value greater than 0 in the new â€˜Aloneâ€™ column indicates that the passenger was not alone. We will use this column to define and analyze the â€˜Aloneâ€™ statistics more clearly.

```python
titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'No'
titanic_df['Alone'].loc[titanic_df['Alone']==0] = 'Yes'
```

```python
titanic_df.head()
```

```python
# Improved Countplot for 'Alone' in Titanic dataset
plt.figure(figsize=(8, 6))

# Create the countplot with enhanced aesthetics
sns.catplot(
    x='Alone',
    data=titanic_df,
    kind='count',
    palette='coolwarm',  # Softer and clearer color palette
    height=6,
    aspect=1.5
)

# Add title and axis labels
plt.title('Distribution of Alone vs. Not Alone Passengers', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Alone', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='darkblue')

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for better presentation
plt.tight_layout()

# Show plot
plt.show()
```

The information is now easier to interpret. Most passengers had family aboard the ship. But did this factor affect their chances of survival? Weâ€™ll explore this question in the next section

## What factors helped someone survive the sinking?


Looking at the informationâ€¦

```python
# Improved Pointplot for 'Pclass' vs 'Survived' in Titanic dataset
plt.figure(figsize=(10, 6))

# Create the pointplot with enhanced aesthetics
sns.catplot(
    x='Pclass',
    y='Survived',
    data=titanic_df,
    kind='point',
    palette='Set2',  # Softer color palette for better visuals
    height=6,
    aspect=1.5
)

# Add title and axis labels
plt.title('Survival Rate by Pclass', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Pclass', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Survived', fontsize=14, fontweight='bold', color='darkblue')

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for better presentation
plt.tight_layout()

# Show plot
plt.show()
```

Not surprisingly, it appears that lower-class passengers had a lower chance of survival.

However, we previously observed that there were more men than women or children in the third class compared to other classes. You might recall the â€˜Women and children first!â€™ protocol from the movie Titanic, where the shipâ€™s crew prioritized women and children when loading lifeboats.

Letâ€™s see if this was truly the case by using the â€˜Personâ€™ column as the hue in the factorplot

```python
# Improved Pointplot for 'Pclass' vs 'Survived' with 'Person' hue
plt.figure(figsize=(10, 6))

# Create the pointplot with enhanced aesthetics
sns.catplot(
    x='Pclass',
    y='Survived',
    hue='Person',
    data=titanic_df,
    kind='point',
    palette='Set1',  # Softer and clearer color palette for differentiation
    height=6,
    aspect=1.5
)

# Add title and axis labels
plt.title('Survival Rate by Pclass and Person', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Pclass', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Survived', fontsize=14, fontweight='bold', color='darkblue')

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for better presentation
plt.tight_layout()

# Show plot
plt.show()
```

This confirms our suspicion: men had lower chances of survival than women and children, as they were given lower priority during the rescue process.

However, we can still conclude that men, women, and children in third class had a lower chance of survival compared to their counterparts in first class.

Now, letâ€™s explore if thereâ€™s a relationship between age and survival.

If you had someone with you on the ship, your chances of survival were higher compared to being alone. It seems that company really does make a difference.
