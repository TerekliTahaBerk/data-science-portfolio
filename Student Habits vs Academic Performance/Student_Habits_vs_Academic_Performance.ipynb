# **Exploring the Impact of Student Habits on Academic Performance**

This notebook embarks on a journey to investigate the relationship between various student habits and their academic performance, specifically focusing on **exam scores**. By analyzing a dataset containing information about study habits, sleep patterns, social media usage, and exercise frequency, we seek to uncover potential correlations and gain insights into factors that may contribute to academic success. Through data visualization and statistical analysis, we hope to identify *key habits* that are associated with higher exam scores, providing valuable information for students and educators alike.

---

## **Setting the Stage:** Importing Libraries and Loading the Dataset

In this initial step, we equip ourselves with the necessary tools for our analysis by importing essential libraries:

* **Pandas:** For data manipulation and analysis.
* **NumPy:** For numerical computations.
* **Matplotlib:** For creating visualizations.
* **Seaborn:** For enhanced statistical graphics.

We then load our dataset containing student habits and exam scores into a Pandas DataFrame, allowing us to easily access and analyze the data.  To get a glimpse of the data's structure, we display the first few rows of the DataFrame. This sets the stage for our exploration into the factors influencing student performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

df = pd.read_csv("/content/student_habits_performance.csv")

df.head()
```

## **Unveiling the Data:** Exploring the Dataset's Structure and Characteristics

Here, we delve deeper into understanding our dataset by examining its structure and key characteristics:

1. **Dataset Size:** We determine the dataset's scope by checking the number of rows and columns.
2. **Data Types:** We investigate the data types of each column, ensuring data integrity and identifying potential areas for preprocessing.
3. **Missing Values and Descriptive Statistics:** By analyzing missing values and calculating descriptive statistics for numerical variables, we gain a comprehensive understanding of the data's distribution and potential outliers.

This thorough exploration lays a solid foundation for further analysis.

```python
print("Shape of dataset: ", df.shape)
df.info()

print("\nMissing values:\n")
print(df.isnull().sum())

df.describe()
```

---

## **Visualizing Exam Score Distribution:** Unveiling the Overall Performance Landscape

This cell focuses on visualizing the distribution of exam scores to gain insights into the overall performance of students:

* **Histogram:** We create a histogram to display the frequency of different score ranges, providing a clear picture of the score distribution.
* **Kernel Density Estimate:** By overlaying a kernel density estimate, we further refine our understanding of the data's shape and density.

This visualization helps us identify trends in exam performance and detect potential areas of concern or excellence.

```python
sns.histplot(df['exam_score'], kde=True, bins=20)
plt.title("Distrubution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

## **Unveiling Potential Connections:** Exploring Relationships between Habits and Exam Scores

In this step, we embark on a journey to uncover potential relationships between student habits and their exam scores:

* **Pairwise Visualizations:** By creating pairwise visualizations of different habits and exam scores, we aim to identify patterns and correlations.

This exploration allows us to generate hypotheses and guide further analysis into the factors that significantly influence exam outcomes.

```python
sns.pairplot(df[['exam_score', 'study_hours_per_day', 'sleep_hours', 'social_media_hours', 'exercise_frequency']])
plt.suptitle("Pairplot: Habits vs Exam Score", y=1.02)
plt.show()
```

## **Delving Deeper into Study Habits:** Investigating the Impact of Daily Study Hours on Exam Performance

Here, we focus specifically on the relationship between daily study hours and exam scores:

* **Box Plot:** Using a box plot, we visualize the distribution of exam scores for different study hour groups, enabling us to compare performance across varying study routines.

This analysis provides valuable insights into the effectiveness of different study approaches and their potential influence on exam outcomes.

```python
sns.boxplot(x='study_hours_per_day', y='exam_score', data=df)
plt.title("Study Hours per Day vs Exam Score")
plt.xlabel("Study Hours per Day")
plt.ylabel("Exam Score")
plt.show()
```

## **Unraveling the Role of Sleep:** Examining the Relationship between Sleep Duration and Exam Performance

This cell investigates the relationship between sleep duration and exam scores, recognizing the importance of sleep for cognitive function and academic success:

* **Scatter Plot with Regression Line:** By plotting a scatter plot with a regression line, we visualize the correlation between sleep hours and exam scores.

This analysis sheds light on the potential impact of sleep patterns on student performance and emphasizes the importance of adequate rest.

```python
sns.lmplot(x="sleep_hours", y="exam_score", data=df)
plt.title("Sleep Hours vs Exam Score")
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Score")
plt.show()
```

## **Unveiling the Network of Relationships:** Identifying Correlations between Variables

In this final step, we aim to identify and quantify the correlations between all variables:

* **Correlation Matrix and Heatmap:** By calculating a correlation matrix and visualizing it as a heatmap, we clearly illustrate the strength and direction of these relationships.

This analysis helps us prioritize the most influential factors affecting exam performance and provides a holistic understanding of the interplay between student habits and academic outcomes.

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Student Habits and Exam Score")
plt.show()
```
